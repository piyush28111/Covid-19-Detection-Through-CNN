```python
import pandas as pd
import os
import shutil
```

```python
# Create the data for positive samples and taking only Covid positive image with PA view of the lung.

File_Path = "covid-chestxray/metadata.csv"
Images_Path = "covid-chestxray/images"
```

```python
df = pd.read_csv(File_Path)
df.hea
```

```python
Target_dir = "DATA/Covid"

if not os.path.exists(Target_dir):
    os.mkdir(Target_dir)
    print('Covid folder created')
```

```python
df = df[df['finding'].str.contains('COVID-19')]

count = 0

for (i,row) in df.iterrows():
    if row['view']=='PA':
        filename = row['filename']
        image_path = os.path.join(Images_Path, filename)
        image_copy_path = os.path.join(Target_dir, filename)
        shutil.copy2(image_path, image_copy_path)
        count+=1
print(count)        
```
